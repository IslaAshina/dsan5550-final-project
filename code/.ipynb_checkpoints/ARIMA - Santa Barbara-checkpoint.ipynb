```python
import pandas as pd

# Load the data from the uploaded CSV file
file_path = '../data/Water Level - Santa Barbara - v2.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
data.head(), data.dtypes
```

```python
# Convert 'Date' and 'Time (GMT)' into a single datetime column
data['Datetime'] = pd.to_datetime(data['Date'] + ' ' + data['Time (GMT)'])

# Extract the hour from the datetime for grouping
data['Hour'] = data['Datetime'].dt.hour
```

### Step 1: ARIMA Model

Let's begin with an ARIMA model. This involves:

- Checking the stationarity of the dataset.
- Determining the parameters for the ARIMA model (p, d, q).
- Training the model on the data.
- Evaluating it on the most recent data.
- Forecasting future values.

First, I will plot the series and perform a test for stationarity to determine if differencing is needed, which is critical for selecting the ARIMA parameters. Let's start by plotting the water level data over time to visualize any trends or seasonality.


```python
import matplotlib.pyplot as plt

# Convert 'Datetime' to datetime type if not already
data['Datetime'] = pd.to_datetime(data['Datetime'])

# Set 'Datetime' as index
data.set_index('Datetime', inplace=True)

# Plotting the verified water levels
plt.figure(figsize=(14, 6))
plt.plot(data['Verified (ft)'], label='Verified Water Level (ft)')
plt.title('Verified Water Level Over Time')
plt.xlabel('Date and Time')
plt.ylabel('Water Level (ft)')
plt.legend()
plt.show()
```

The plot shows how the verified water levels fluctuate over time. To proceed with ARIMA, we need to assess the stationarity of the time series because ARIMA requires the series to be stationary. We'll use the Augmented Dickey-Fuller (ADF) test to check for stationarity.

### Step 2: Testing for Stationarity
The ADF test will help us determine whether the series is stationary or if we need to perform differencing as part of setting up the ARIMA model. Let's perform this test now.

```python
from statsmodels.tsa.stattools import adfuller

# Perform Augmented Dickey-Fuller test to check stationarity
adf_result = adfuller(data['Verified (ft)'].dropna())

# Display the ADF statistic and p-value
adf_statistic, p_value = adf_result[0], adf_result[1]
adf_statistic, p_value
```

The ADF test results show a statistic of -2.816 and a p-value of 0.056. Typically, a p-value less than 0.05 suggests the series is stationary. Here, the p-value is slightly above this threshold, indicating the series might not be stationary. To ensure robustness in our ARIMA model, we should consider differencing the series.

### Step 3: Differencing the Series
We'll apply a first difference to the series and then retest for stationarity. Let's do this next

```python
# Apply first difference
data['Verified_diff'] = data['Verified (ft)'].diff()

# Plot the differenced data
plt.figure(figsize=(14, 6))
plt.plot(data['Verified_diff'], label='First Difference of Verified Water Level (ft)')
plt.title('First Difference of Verified Water Level')
plt.xlabel('Date and Time')
plt.ylabel('Difference in Water Level (ft)')
plt.legend()
plt.show()

# Re-run the ADF test on the differenced data
adf_result_diff = adfuller(data['Verified_diff'].dropna())
adf_statistic_diff, p_value_diff = adf_result_diff[0], adf_result_diff[1]
adf_statistic_diff, p_value_diff
```

The differenced series shows a much more stationary behavior, as evidenced by the plot. The ADF test on this differenced data gives a statistic of -33.43 and a p-value of 0.0, strongly indicating stationarity.

### Step 4: Building the ARIMA Model
Now that the series is confirmed stationary, we can configure and fit an ARIMA model. We'll use the (p, d, q) parameters where `p` and `q` will initially be set based on typical starting points for such models, and `d` is 1 since we applied first differencing.

Let’s fit an ARIMA model to the differenced data and evaluate its performance.

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit the ARIMA model
model = ARIMA(data['Verified (ft)'], order=(1, 1, 1))  # Using 1,1,1 for p,d,q
arima_result = model.fit()

# Display the summary of the ARIMA model
arima_result.summary()
```

```python
# Forecast the next 24 hours
forecast_result = arima_result.get_forecast(steps=24)
forecast_mean = forecast_result.predicted_mean
forecast_conf_int = forecast_result.conf_int()
```

```python
# Plotting the forecast without the confidence interval due to the previous error
plt.figure(figsize=(14, 6))
plt.plot(data['Verified (ft)'].iloc[-168:], label='Historical Verified Water Level')
plt.plot(forecast_mean.index, forecast_mean, color='red', label='Forecasted Water Level')
plt.title('ARIMA Forecast of Verified Water Level')
plt.xlabel('Date and Time')
plt.ylabel('Water Level (ft)')
plt.legend()
plt.show()
```
