```python
import pandas as pd

# Load the dataset
file_path = '../data/Water Level - Santa Barbara - v2.csv'
data = pd.read_csv(file_path)

# Display the first few rows and the data info
data.head(), data.info(), data.describe()
```

```python
# Combine Date and Time into a single datetime column and set as index
data['Datetime'] = pd.to_datetime(data['Date'] + ' ' + data['Time (GMT)'])
data.set_index('Datetime', inplace=True)

# Drop the original Date and Time columns
data.drop(['Date', 'Time (GMT)'], axis=1, inplace=True)

# Plotting the data
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Predicted (ft)'], label='Predicted (ft)', alpha=0.6)
plt.plot(data.index, data['Verified (ft)'], label='Verified (ft)', alpha=0.6)
plt.title('Water Levels over Time')
plt.xlabel('Date')
plt.ylabel('Water Level (ft)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Extract time-based features
data['hour'] = data.index.hour
data['day_of_week'] = data.index.dayofweek
data['month'] = data.index.month

# Display the updated DataFrame to check the new features
data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))

# Scale the 'Verified (ft)' column
data['Verified_scaled'] = scaler.fit_transform(data[['Verified (ft)']])

# Display some of the scaled data to confirm scaling
data[['Verified (ft)', 'Verified_scaled']].head()
```

```python
import numpy as np

# Function to create sequences
def create_sequences(data, sequence_length):
    X = []
    y = []
    for i in range(len(data) - sequence_length):
        X.append(data.iloc[i:(i + sequence_length), -1].values)  # Get only the scaled verified values
        y.append(data.iloc[i + sequence_length, -1])  # Target is the next hour verified value
    return np.array(X), np.array(y)

# Define the sequence length (24 hours)
sequence_length = 24

# Generate sequences and targets
X, y = create_sequences(data, sequence_length)

# Display shapes of the created arrays
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
train_size = int(len(X) * 0.9)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Display the shape of training and testing data
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional
from tensorflow.keras.optimizers import Adam, RMSprop, SGD
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Define the LSTM model architecture
model = Sequential([
    LSTM(50, input_shape=(24, 1)),  # 50 LSTM units, input_shape is (sequence_length, number of features)
    Dense(1)  # Output layer to predict the next value
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Display model summary
model.summary()
```

##### Tuning Case 3

```python
!pip install -q -U keras-tuner
import kerastuner as kt
import tensorflow as tf

def build_model(hp):
    model = Sequential()
    model.add(Bidirectional(
        LSTM(units=hp.Int('units', min_value=32, max_value=512, step=32), 
             return_sequences=True), 
             input_shape=(24, 1)))
    model.add(Dropout(rate=hp.Float('dropout_1', min_value=0.0, max_value=0.5, step=0.1)))
    model.add(LSTM(units=hp.Int('units', min_value=32, max_value=512, step=32)))
    model.add(Dropout(rate=hp.Float('dropout_2', min_value=0.0, max_value=0.5, step=0.1)))
    model.add(Dense(1))
    
    model.compile(
        optimizer=hp.Choice('optimizer', values=['adam', 'rmsprop', 'sgd']),
        loss='mean_squared_error')
    
    return model

tuner = kt.Hyperband(
    build_model,
    objective='val_loss',
    max_epochs=50,
    hyperband_iterations=2)

tuner.search(X_train, y_train, epochs=50, validation_data=(X_test, y_test), callbacks=[tf.keras.callbacks.EarlyStopping('val_loss', patience=5)])
```

```python
# Retrieve the best hyperparameters
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

# Build the model with the best hyperparameters
model = build_model(best_hps)

# Display model summary to confirm the architecture
model.summary()
```

```python
# You can retrieve specific hyperparameter values as follows
best_learning_rate = best_hps.get('learning_rate') if 'learning_rate' in best_hps else 0.001  # default value if not tuned
best_optimizer = best_hps.get('optimizer')

# Compile the model with the optimizer chosen by the tuner
model.compile(
    optimizer=best_optimizer,
    loss='mean_squared_error'
)

# Display model summary to confirm the architecture
model.summary()
```

##### Tuning Case 4

```python
from tensorflow.keras.optimizers import Adam, RMSprop, SGD

def build_model(hp):
    model = Sequential()
    model.add(Bidirectional(
        LSTM(units=hp.Int('units', min_value=32, max_value=512, step=32), 
             return_sequences=True), 
             input_shape=(24, 1)))
    model.add(Dropout(rate=hp.Float('dropout_1', min_value=0.0, max_value=0.5, step=0.1)))
    model.add(LSTM(units=hp.Int('units', min_value=32, max_value=512, step=32)))
    model.add(Dropout(rate=hp.Float('dropout_2', min_value=0.0, max_value=0.5, step=0.1)))
    model.add(Dense(1))
    
    # Tuning the batch size
    batch_size = hp.Int('batch_size', min_value=32, max_value=256, step=32)
    
    # Tuning the learning rate
    lr = hp.Float('learning_rate', min_value=1e-4, max_value=1e-2, sampling='log')

    # Tuning the optimizer
    optimizer_name = hp.Choice('optimizer', values=['adam', 'rmsprop', 'sgd'])
    if optimizer_name == 'adam':
        optimizer = Adam(learning_rate=lr)
    elif optimizer_name == 'rmsprop':
        optimizer = RMSprop(learning_rate=lr)
    elif optimizer_name == 'sgd':
        optimizer = SGD(learning_rate=lr)

    model.compile(optimizer=optimizer, loss='mean_squared_error')
    
    return model
```

```python
# Define the tuner
tuner = kt.Hyperband(
    build_model,
    objective='val_loss',
    max_epochs=50,
    hyperband_iterations=2,
    directory='my_dir',  # Specify a directory to store the hyperband files
    project_name='hyperband_tuning'
)

# Search for the best hyperparameters including batch_size
tuner.search(
    X_train, y_train,
    epochs=50,  # This should be the maximum number of epochs to train
    validation_data=(X_test, y_test),
    callbacks=[tf.keras.callbacks.EarlyStopping('val_loss', patience=10)]
)
```

```python
# Retrieve the best hyperparameters
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
best_batch_size = best_hps.get('batch_size')

# Build the model with the best hyperparameters
model = build_model(best_hps)

# No need to compile again since it's already done within the `build_model` function
```

```python
# Display model summary to confirm the architecture
model.summary()
```

### Insights from Evaluation Metrics:
- Mean Absolute Error (MAE) ~ 0.564: This indicates that on average, the predictions are approximately 0.594 feet off from the true water levels. Considering the range of water levels, this error might be considered moderate. The importance of this error depends on the context and how these predictions are used.
- Mean Squared Error (MSE) ~ 0.489: MSE is more sensitive to outliers than MAE because it squares the errors. A value of 0.538 tells us that the errors are somewhat significant, but without knowing the context of what water level deviations are critical, it's hard to judge the severity.
- Root Mean Squared Error (RMSE) ~ 0.699: RMSE gives us an error magnitude that is in the same units as the water level measurements (feet), making it more interpretable than MSE. An RMSE of 0.734 feet suggests that the model's predictions are reasonably close to the actual measurements, though there is room for improvement.
- R-squared ~ 0.781: This value indicates that approximately 75.9% of the variance in the true water levels is captured by the model's predictions. This suggests that the model has a good predictive power, although there's about 24.1% of the variance that the model does not capture.
