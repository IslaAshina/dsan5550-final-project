```python
import pandas as pd

# Load the uploaded Excel file to examine its content
file_path = '../data/model performance.xlsx'
data = pd.read_excel(file_path)
data
```

```python
import matplotlib.pyplot as plt

# Setting up the figure and axes
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# Bar plot for RMSE
axes[0].bar(data['Model Name'], data['RMSE'], color='skyblue')
axes[0].set_title('RMSE by Model')
axes[0].set_ylabel('RMSE')
axes[0].set_xlabel('Model')
axes[0].set_xticklabels(data['Model Name'], rotation=45, ha='right')

# Bar plot for R-squared
axes[1].bar(data['Model Name'], data['R-squared'], color='salmon')
axes[1].set_title('R-squared by Model')
axes[1].set_ylabel('R-squared')
axes[1].set_xlabel('Model')
axes[1].set_xticklabels(data['Model Name'], rotation=45, ha='right')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
