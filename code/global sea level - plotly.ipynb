```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Read the data into a Pandas DataFrame
data = pd.read_csv('../data/sea_levels_2015.csv')

# Convert the 'Time' column to datetime format
data['Time'] = pd.to_datetime(data['Time'], format='%Y-%m-%d')

# Set the 'Time' column as the index
data = data.set_index('Time')

# Create the time series plot
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(data.index, data['GMSL'])

# Customize the plot
ax.set_title('Global Mean Sea Level (GMSL)')
ax.set_xlabel('Time')
ax.set_ylabel('GMSL (mm)')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Set the x-axis ticks to show years
years = mdates.YearLocator()
years_fmt = mdates.DateFormatter('%Y')
ax.xaxis.set_major_locator(years)
ax.xaxis.set_major_formatter(years_fmt)

# Show the plot
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
from datetime import datetime

# Read the data into a Pandas DataFrame
data = pd.read_csv('../data/sea_levels_2015.csv')

# Convert the 'Time' column to datetime format
data['Time'] = pd.to_datetime(data['Time'], format='%Y-%m-%d')

# Set the 'Time' column as the index
data = data.set_index('Time')

# Create the time series plot
fig = go.Figure()
fig.add_trace(go.Scatter(x=data.index, y=data['GMSL'], mode='lines'))

# Customize the plot
fig.update_layout(
    title='Global Mean Sea Level (GMSL)',
    xaxis_title='Time',
    yaxis_title='GMSL (mm)',
    xaxis=dict(
        tickmode='array',
        tickvals=[datetime(year, 1, 1) for year in range(1870, 2021, 10)],
        ticktext=[str(year)[:4] + 's' for year in range(1870, 2021, 10)]
    )
)

# Show the plot
fig.show()
```
